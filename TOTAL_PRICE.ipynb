```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("olx_new_data.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = data.select_dtypes(include=['object']).columns

# Apply label encoding to each categorical column
label_encoder = LabelEncoder()
for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data.info()
```

```python
data.drop(columns='Address',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X =data.drop(['Total Price (INR)'],axis=1)
y = data['Total Price (INR)'] 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(16,9))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True)
```

```python
train_data['Carpet Area']=np.log(train_data['Carpet Area']+1)
train_data['Floor No']=np.log(train_data['Floor No']+1)
train_data['SquareFeet']=np.log(train_data['SquareFeet']+1)
```

```python
train_data.hist(figsize=(16,9))
```

```python

train_data['Bedroom']=np.log(train_data['Bedroom']+1)
train_data['Bathroom']=np.log(train_data['Bathroom']+1)
train_data['Furnished Status']=np.log(train_data['Furnished Status']+1)
train_data['Construction Status']=np.log(train_data['Construction Status']+1)
train_data['Car Parking']=np.log(train_data['Car Parking']+1)
train_data['BHK']=np.log(train_data['BHK']+1)
train_data['State']=np.log(train_data['State']+1)
```

```python
train_data.hist(figsize=(16,9))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train,y_train = train_data.drop(['Total Price (INR)'],axis=1),train_data['Total Price (INR)']
x_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()
reg.fit(x_train_s,y_train)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train_s,y_train)
```

```python
forest.score(X_test_s,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth":[None,4,8]
}
grid_search = GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score = True)
grid_search.fit(x_train_s,y_train)
```

```python
best_forest = grid_search.best_estimator_
best_forest.score(X_test_s,y_test)
```
